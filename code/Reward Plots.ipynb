## Curriculum Learning (Original)

```python
import re
import pandas as pd
# Loading the data from the file
file_path = 'file5_1.txt'
data = pd.read_csv(file_path, header=None)

# Displaying the first few rows of the data to understand its structure and format
data.head()
# Extracting the average episode reward values
reward_pattern = re.compile(r'\|\s+ep_rew_mean\s+\|\s+([-\d.]+)\s+\|')
rewards = data[0].str.extract(reward_pattern).dropna().astype(float).squeeze()

# Showing the first few extracted reward values
rewards.head()
```

```python
import re
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculating the running average and standard deviation
window_size = 100  # Window size for running average and std calculation
running_avg = rewards.rolling(window=window_size, min_periods=1).mean()
running_std = rewards.rolling(window=window_size, min_periods=1).std()

# Plotting the reward values, running average, and standard deviation
plt.figure(figsize=(12, 6))

# Plotting the reward values
plt.plot(rewards.index, rewards.values, label='Reward', alpha=0.6)

# Plotting the running average
# plt.plot(rewards.index, running_avg.values, label=f'Running Average (window size = {window_size})', linewidth=2)

# Plotting the standard deviation
plt.fill_between(rewards.index, rewards.values - running_std, rewards.values + running_std, alpha=0.2, label='Std Dev')

# Adding labels and title
plt.xlabel('Training Steps')
plt.ylabel('Reward')
plt.title('Reward Plot with Running Average and Standard Deviation')
plt.legend()
plt.grid(True)
plt.xlim(48500, 60000)
plt.show()
```

```python
file_path = 'file5_7.txt'
data = pd.read_csv(file_path, header=None)
reward_pattern = re.compile(r'\|\s+ep_rew_mean\s+\|\s+([-\d.]+)\s+\|')
rewards = data[0].str.extract(reward_pattern).dropna().astype(float).squeeze()

window_size = 100  # Window size for running average and std calculation
running_avg = rewards.rolling(window=window_size, min_periods=1).mean()
running_std = rewards.rolling(window=window_size, min_periods=1).std()

plt.figure(figsize=(12, 6))

plt.plot(rewards.index, rewards.values, label='Reward', alpha=0.6)

plt.fill_between(rewards.index, rewards.values - running_std, rewards.values + running_std, alpha=0.2, label='Std Dev')

plt.xlabel('Training Steps')
plt.ylabel('Reward')
plt.title('Reward Plot with Running Average and Standard Deviation')
plt.legend()
plt.grid(True)
# plt.xlim(0, 14000)
plt.show()
```

```python
difficulty_pattern = re.compile(r'Training on difficulty level (\d+)|Currently Difficulty Level (\d+)')
difficulties = data[0].str.extract(difficulty_pattern).fillna(method='ffill').fillna(1).astype(int).max(axis=1)

# Initialize variables
current_difficulty = 1
rewards = []
difficulties = []

# Read the file line by line and extract the difficulty levels and rewards
with open(file_path, 'r') as file:
    for line in file:
        # Check for difficulty level
        difficulty_match = difficulty_pattern.search(line)
        if difficulty_match:
            group1, group2 = difficulty_match.groups()
            current_difficulty = int(group1) if group1 is not None else int(group2)
        
        # Check for reward
        reward_match = reward_pattern.search(line)
        if reward_match:
            rewards.append(float(reward_match.group(1)))
            difficulties.append(current_difficulty)

# Create a DataFrame to hold the extracted data
combined_data = pd.DataFrame({'Reward': rewards, 'Difficulty': difficulties})

combined_data.head()
```

```python
window_size = 100  # Window size for running average and std calculation

# Calculate the running average and standard deviation for each difficulty level
combined_data['Running Avg'] = combined_data.groupby('Difficulty')['Reward'].transform(lambda x: x.rolling(window=window_size, min_periods=1).mean())
combined_data['Running Std'] = combined_data.groupby('Difficulty')['Reward'].transform(lambda x: x.rolling(window=window_size, min_periods=1).std())

# Plot for difficulty level 1
data_level_1 = combined_data[combined_data['Difficulty'] == 1]

# Plot for difficulty level 2
data_level_2 = combined_data[combined_data['Difficulty'] == 2]

# Plot for difficulty level 3
data_level_3 = combined_data[combined_data['Difficulty'] == 3]

# Plot the data with legends for standard deviation
plt.figure(figsize=(12, 6))

# Plot for difficulty level 1
plt.plot(data_level_1.index, data_level_1['Reward'], label='Reward Level 1', alpha=0.6)
plt.fill_between(data_level_1.index, data_level_1['Reward'] - data_level_1['Running Std'], data_level_1['Reward'] + data_level_1['Running Std'], alpha=0.2, label='Std Dev Level 1')

# Plot for difficulty level 2
plt.plot(data_level_2.index, data_level_2['Reward'], label='Reward Level 2', alpha=0.6)
plt.fill_between(data_level_2.index, data_level_2['Reward'] - data_level_2['Running Std'], data_level_2['Reward'] + data_level_2['Running Std'], alpha=0.2, label='Std Dev Level 2')

# Plot for difficulty level 3
plt.plot(data_level_3.index, data_level_3['Reward'], label='Reward Level 3', alpha=0.6)
plt.fill_between(data_level_3.index, data_level_3['Reward'] - data_level_3['Running Std'], data_level_3['Reward'] + data_level_3['Running Std'], alpha=0.2, label='Std Dev Level 3')

plt.xlabel('Training Steps')
plt.ylabel('Reward')
plt.title('Reward Plot with Running Average and Standard Deviation for Different Difficulty Levels')
plt.legend()
plt.grid(True)
plt.show()

```
